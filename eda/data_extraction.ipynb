```python
import pandas as pd
import numpy as np
import ast
from pathlib import Path
```

## Bundle Data

```python
fp = Path("data") / "bundle_data.json"
bundle_data = list()

with open(fp, "r", encoding='utf-8') as f:
    for line in f:
        try:
            parsed_dict = ast.literal_eval(line.strip())
            bundle_data.append(parsed_dict)
        except (ValueError, SyntaxError) as e:
            print(f"Error parsing line: {e}")

print(f"Successfully parsed {len(bundle_data)} records")
bundle_data[0]
```

```python
bundle = pd.DataFrame(bundle_data)
bundle.head(3)
```

```python
save_fp = Path("data") / "bundle.csv"
bundle.to_csv(save_fp, index=False)
```
